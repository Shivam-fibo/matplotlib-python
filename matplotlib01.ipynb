```python
# from the pandas dataframe
batsman= pd.read_csv('sharma-kohli.csv')
batsman
```

```python
 plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
 plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# plotting multilple graph
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
#label
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', linestyle = 'solid',linewidth = 3 )
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dotted')
plt.title("Rohit Sharma V/s Virat Kohli")
plt.ylabel('Runs Scored')
plt.xlabel("Seasons")
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', marker = '.', markersize = 13 )
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black',marker='*', markersize = 13)
plt.title("Rohit Sharma V/s Virat Kohli")
plt.ylabel('Runs Scored')
plt.xlabel("Seasons")
```

```python
# to print lable inside the grph use legned function
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green', label = 'V Kohli' )
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', label = 'RG Sharma')
plt.title("Rohit Sharma V/s Virat Kohli")
plt.ylabel('Runs Scored')
plt.xlabel("Seasons")
plt.legend(loc = 'upper right')
# loc is used to figure out the best position to place this 
```

```python
df = pd.read_csv('batter.csv')
df = df.head(50)
plt.scatter(df['avg'], df['strike_rate'], color = 'red')
plt.title("batsman average vs strike rate")
plt.xlabel('average')
plt.ylabel('strike rate')
```

```python
#without using scatter function we can still plot the graph with the help of plt fucntion itself
plt.plot(df['avg'], df['strike_rate'], 'o')
```

```python
childern = [13,23,15,17,20]
colors = ['red', 'blue', 'green', 'yellow', 'pink']
plt.bar(colors, childern)
plt.title('Colos with children')
plt.xlabel('color')
plt.ylabel('age')
```

```python
# for horizantal barchat
plt.barh(colors, childern)
plt.grid()
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
np.arange(df.shape[0])-0.5
```

```python
plt.bar(np.arange(df.shape[0])-0.2 , df['2015'], width = 0.2)
plt.bar(np.arange(df.shape[0]) , df['2016'], width = 0.2)
plt.bar(np.arange(df.shape[0])+0.2 , df['2017'], width=0.2)
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
# stack bar chart

plt.bar(df['batsman'], df['2015'], label = '2015', color = 'red')
plt.bar(df['batsman'], df['2016'], bottom = df['2015'], label = '2016', color = 'orange')
plt.bar(df['batsman'], df['2016'], bottom = df['2015']+ df['2016'], label = '2017')
plt.legend()
plt.show()
```
